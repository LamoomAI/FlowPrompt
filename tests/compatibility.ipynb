```python
import subprocess
import sys

# Function to create the virtual environment
def create_venv(venv_name):
    subprocess.run([sys.executable, "-m", "venv", venv_name], check=True)
    print(f"Virtual environment '{venv_name}' created.")

venv_name = 'venv_test'

create_venv(venv_name)

# Install required packages
packages = ['pytest', 'python-dotenv']
for package in packages:
    subprocess.run([f"./{venv_name}/Scripts/pip", "install", package], check=True)
```

```python
def reinstall_package_in_venv(venv_name, package, version):
    subprocess.run([f"./{venv_name}/Scripts/pip", "uninstall", "-y", package], check=True)
    subprocess.run([f"./{venv_name}/Scripts/pip", "install", f"{package}=={version}"], check=True)
    print(f"Reinstalled {package} version {version} in {venv_name}")

# Loop through versions we need
versions = ['0.1.29', '0.1.30']
test_files = ['./compatability/test_save_logs.py']

all_output = ""
for version in versions:
    reinstall_package_in_venv(venv_name, 'flow-prompt', version) 

    for test_file in test_files:
        result = subprocess.run(
            [f"./{venv_name}/Scripts/python", "-m", "pytest", test_file, "-v", "-s"],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        
        all_output += f"\n--- Pytest Output for Version {version} and File {test_file} ---\n"
        all_output += result.stdout
        all_output += result.stderr
        all_output += "\n"

print(all_output)
```
